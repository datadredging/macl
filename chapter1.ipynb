Optimizations of numpy

```python
import timeit
import numpy as np

number = 100000
standart = timeit.timeit('sum(x*x for x in range(1000))', number=number)
better = timeit.timeit('sp.sum(x*x)', setup='import scipy as sp; import numpy as np; x=np.arange(1000)', number=number)
best =   timeit.timeit('x.dot(x)', setup='import numpy as np; x=np.arange(1000)', number=number)

print "Standart: %f sec" % standart
print "Better: %f sec" % better
print "Best: %f sec" % best
```

Web traffic analysis

```python
import scipy as sp
data = sp.genfromtxt('web_traffic.tsv', delimiter='\t',)
print data.shape

x = data[:,0]
y = data[:,1]
x = x[~np.isnan(y)]
y = y[~np.isnan(y)]
print y.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
def basic_plot(figsize=(10,5)):
    plt.figure(figsize=figsize)
    plt.scatter(x,y,s=10)
    plt.title('Web traffic over month')
    plt.xlabel('Time')
    plt.ylabel('Hits/hour')
    plt.xticks([w*7*24 for w in range(10)], ['week %i' % i  for i in range(10)])
    plt.autoscale(tight=True)
    plt.grid(True, linestyle = '-', color = '0.75')
basic_plot()
plt.show()
```

```python
def error(f, x, y):
    return np.sum((f(x)-y)**2)

def eq( a, b, eps=0.0001 ):
    return abs(a - b) <= eps
```

Get simple linear model

```python
fp1, residuals, rank, sv, rcond = sp.polyfit(x,y,deg=1,full=True)
print fp1
f1 = sp.poly1d(fp1)
print eq(residuals, error(f1, x, y))
basic_plot()
plt.plot(x,f1(x),linewidth=4)
plt.legend(['d = %i' % f1.order],loc='upper left')
plt.show()
```

More complex one

```python
fp2 = sp.polyfit(x,y,2)
f2 = sp.poly1d(fp2)
basic_plot()
plt.plot(x,f2(x),linewidth=4)
plt.legend(['d = %i' % f2.order],loc='upper left')
plt.show()
```

degree madness

```python
fp10 = sp.polyfit(x,y,10)
f10 = sp.poly1d(fp10)
fp40 = sp.polyfit(x,y,40)
f40 = sp.poly1d(fp40)
fp100 = sp.polyfit(x,y,100)
f100 = sp.poly1d(fp100)
basic_plot(figsize=(15,10))
models = [f1, f2, f10, f40, f100]
plt.plot(x,f1(x),linewidth=4)
plt.plot(x,f2(x),linewidth=4)
plt.plot(x,f10(x),linewidth=4)
plt.plot(x,f40(x),linewidth=4)
plt.plot(x,f100(x),linewidth=4)
plt.legend(['d = %i, error = %f' % (f.order,error(f,x,y)/10000) for f in models],loc='upper left')
plt.show()
```

Data separation


```python
sep_point = 3.5*7*24
xbefore = x[:sep_point] 
ybefore = y[:sep_point]
xafter = x[sep_point:]
yafter = y[sep_point:]

fa = sp.poly1d(sp.polyfit(xafter,yafter,1))
fb = sp.poly1d(sp.polyfit(xbefore,ybefore,1))

basic_plot()
plt.plot(xbefore,fb(xbefore),linewidth=4)
plt.plot(xafter,fa(xafter), linewidth=4, linestyle='-.', c='r')
plt.show()
```

plot future

```python
basic_plot()
xx = np.linspace(0,x[-1]*1.5,1000)
plt.plot(xx,f1(xx),linewidth=4)
plt.plot(xx,f2(xx),linewidth=4)
plt.plot(xx,f10(xx),linewidth=4)
plt.plot(xx,f40(xx),linewidth=4)
plt.plot(xx,f100(xx),linewidth=4)
plt.ylim([1000,6000])
plt.show()
```
